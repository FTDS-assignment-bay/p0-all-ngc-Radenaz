```python
query_pat = """
WITH parsed_data AS (
  SELECT
    PARSE_DATE('%Y%m%d', Grant_or_Publication_Date) AS grant_date,
    COUNT(*) AS patent_count
  FROM `patents-public-data.uspto_oce_cancer.publications`
  WHERE
    CAST(Grant_or_Publication_Date as STRING) BETWEEN '20120101' AND '20231231'
    AND LOWER(Patent_Title) LIKE '%data mining%'
    GROUP BY grant_date
),
cumu_data AS (
  SELECT
    EXTRACT(YEAR FROM grant_date) AS year,
    EXTRACT(MONTH FROM grant_date) AS month,
    SUM(patent_count) OVER (ORDER BY EXTRACT(YEAR FROM grant_date), EXTRACT(MONTH FROM grant_date)) AS total_patents
  FROM parsed_data
)
SELECT
  year,
  month,
  total_patents
FROM cumu_data
ORDER BY year DESC, month ASC
"""

query_pat = client.query(query_pat).to_dataframe()
query_pat
```
