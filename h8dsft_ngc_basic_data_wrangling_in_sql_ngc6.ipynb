# Create a query to answer the following questions:


*   How many tables are in the `austin_bikeshare` dataset?

```python
que_a= """
SELECT COUNT(*) AS table_count
FROM `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.TABLES`
"""
table_count = client.query(que_a).to_dataframe()
table_count
```

*   How many columns in the `bikeshare_stations` table have STRING data?

```python
que_b = """
SELECT COUNT(*) AS columns_string_type
FROM `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS`
WHERE
  table_name = 'bikeshare_stations'
  AND data_type IN ('STRING')
"""
stringtype_columns_df = client.query(que_b).to_dataframe()
stringtype_columns_df
```

# Create a query to filter data based on the following conditions:

*   Retrieve all trips where the bike_id is not null and the duration is greater than 1000 minutes.


```python
que_c = """
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE
  bike_id IS NOT NULL
  AND duration_minutes > 1000
"""
trips_conditions = client.query(que_c).to_dataframe()
trips_conditions
```

*   Filter trips where the `subcriber_type` (NOT `trip_id`) is `"Walk Up"` and the `end_station_name`  is "Stolen".

```python
query_fixed = """
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE
  subscriber_type = 'Walk Up'
  AND end_station_name = 'Stolen'
"""
trips_filtered = client.query(query_fixed).to_dataframe()
trips_filtered
```
